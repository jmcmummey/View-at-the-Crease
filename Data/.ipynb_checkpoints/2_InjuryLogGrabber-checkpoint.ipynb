<h1>Get the player injury logs</h1>

```python
import pandas as pd
import numpy as np
import requests as r
import re
from bs4 import BeautifulSoup as bs

pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
```

```python
#import player list from csv and tidy 
players = pd.read_csv('nhlplayerlist.txt','\t')
players.drop(columns=['Unnamed: 0','index'],axis=1,inplace=True)
players.head(5)
```

```python
#website - check for Craig Anderson
url= 'http://www.prosportstransactions.com/hockey/Search/SearchResults.php?Player=Craig+Anderson&Team=&BeginDate=1990-09-01&EndDate=&ILChkBx=yes&InjuriesChkBx=yes&submit=Search&start=0'
page = r.get(url)
soup = bs(page.text,'html.parser')

injurytable = pd.DataFrame()
table = soup.find_all('tr')

for ri,row in enumerate(table):
    rowe = []
    for c in row.find_all('td'):
        rowe.append(c.text)
    if ri==0:
        injurytable = pd.DataFrame(columns=rowe)
    else:
        injurytable = injurytable.append(dict(zip(injurytable.columns,rowe)),ignore_index=True)

injurytable
```
