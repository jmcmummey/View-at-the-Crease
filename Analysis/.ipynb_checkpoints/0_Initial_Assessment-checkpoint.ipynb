<p>The purpose of this sheet is to look for trends in the data might be useful in the model.  This was performed on hockey players from all positions.</p>

```python
import pandas as pd
import numpy as np
import os
from datetime import datetime as dt, timedelta
from dateutil.relativedelta import *
import re
from matplotlib import pyplot as plt
import sqlite3
```

<h3>Player Data</h3>

```python
#import player list data
players = pd.read_csv('..\\Data\\nhlplayerlistr3.txt','\t')
players.drop(columns=['Unnamed: 0','link','height','weight'],inplace=True)
players = players.set_index('unique_id')
#sort based ons whether or not there is injury data
players.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.figsize = (10,10)
fig.dpi = 150
fig.subplots_adjust(wspace=0.5)
ax1.plot(players['weight_kg'],players['injuries_found'],'r.')
ax1.set_xlabel('weight')
ax1.set_ylabel('No of injuries')
ax1.set_title('Injuries vs Player Height')
ax2.plot(players['height_cm'],players['injuries_found'],'r.')
ax2.set_xlabel('Height')
ax2.set_ylabel('No of injuries')
ax2.set_title('Injuries vs Player Height')
plt.show()
```

```python
print("Looking at the impact of weight on injuries, theres an increase in the weighted mean of weight to %0.2f vs Mean %0.2f"
      %((players['weight_kg']*players['injuries_found']).sum()/(players['injuries_found']).sum(),players['weight_kg'].mean()))
```

```python
print("Looking at the impact of height on injuries, theres an increase in the weighted mean of height to %0.2f vs Mean %0.2f"
      %((players['height_cm']*players['injuries_found']).sum()/(players['injuries_found']).sum(),players['height_cm'].mean()))
```
